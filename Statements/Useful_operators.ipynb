#### Range : Function that allows to quickly generate a list of integers

```python
range(0,10)
```

```python
# Cast it to list
list(range(0,11))
```

```python
list1 = list(range(0, 21))

for i in list1:
    if i % 2 == 0:
        print(f"The number divisible by 2 is {i}")
```

### Enumerate

```python
idx_count = 0

for l1 in 'Rohan':
  print(f'At index {idx_count} the letter is {l1}')
  idx_count = idx_count + 1
```

```python
for i,l in enumerate('Rohan'):
  print(f'At index {i} the letter is {l}')
```

```python
list(enumerate('Rohan'))
```

### Zip

```python
l1 = [21,22,23,24,25]
l2 = ['a','b','c','d','e']
list(zip(l1,l2))
```

```python
for i1, i2 in zip(l1,l2):
  print(f'For this tuple, first item was {i1}, and second item was {i2}')
```

### In

```python


'a' in ['hi',12,'world',34.9,['shreyas','hello'],{'k1':'value1'}]
```

```python
['shreyas','hello'] in ['hi',12,'world',34.9,['shreyas','hello'],{'k1':'value1'}]
```

### Not in

```python
#not in

'a' not in ['hi',12,'world',34.9,['shreyas','hello'],{'k1':'value1'}]
```
