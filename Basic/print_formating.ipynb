### Formatting with percentage

* %s for string and number
* %r for raw text (if you want it as it is with text)
* %f for float no.

```python
#% = String formatting operator

name = input("Enter your name:")
print("Hello %s, Welcome to this repo!" %name)
```

```python
name = input("Enter your name:")
age = input("Enter your age:")
print("Hi my name is %s and my age is %s." %(name.title(),age))
```

### Padding and precision of floating point numbers
* %5.2f : This mean 5 is the min no. of the characters the string should have and .2 is the how many number should be visible after the decimal ppoint

```python
print("My floating point number is %2.2f" %(25.2456))
```

```python
print("My floating point number is %30.2f" %(25.2456))
```
